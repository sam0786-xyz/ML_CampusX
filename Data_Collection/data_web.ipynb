```python
#Use selenium if beatifulsoup does not work, because of dynamic websites
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd

# Setup Chrome driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Open Naukri Internship URL
url = "https://www.naukri.com/machine-learning-intern-internship-jobs-in-delhi-ncr"
driver.get(url)

# Wait for JS content to load
time.sleep(5)  # Use WebDriverWait later for smarter loading

# Find all job cards with given full class
job_cards = driver.find_elements(
    By.XPATH,
    "//div[contains(@class, 'cust-job-tuple') and contains(@class, 'internship-tuple')]"
)

job_title = []
company_name = []

# Loop through job listings
for job in job_cards:
    try:
        title = job.find_element(By.CLASS_NAME, "title").text.strip()
    except:
        title = "N/A"

    try:
        company = job.find_element(By.CLASS_NAME, "comp-name").text.strip()
    except:
        company = "N/A"

    job_title.append(title)
    company_name.append(company)

# Store in DataFrame
df = pd.DataFrame({
    'Job Title': job_title,
    'Company': company_name
})

# Output first few rows
print(df.head())

# Save to CSV
df.to_csv("naukri_internships.csv", index=False)

# Close browser
driver.quit()
```
